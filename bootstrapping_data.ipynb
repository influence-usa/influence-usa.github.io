```python
with open('data/campaign_finance.json','r') as fin:
    jd = json.load(fin)
    new_records = []
    for record in jd:
        new_record = deepcopy(record)
        jurisdiction = record.pop('jurisdiction')
        if jurisdiction not in ['US Federal',]:
            disclosure_type = record.pop('disclosuretype')
            difficulty = record.pop('difficulty')
            for task, status in record.items():
                if len(status) > 0:
                    if status == "0":
                        issue_title = "{t} for {j} {dt}".format(
                                t=task, j=jurisdiction, dt=disclosure_type)
                        res = gh.search_issues('repo:influence-usa/scrapers-us-state "{t}"'.format(
                            t=issue_title))
                        if res.totalCount == 0:
                            print("create '{}'".format(issue_title))
                            new_issue = repo.create_issue(issue_title)
                            new_record[task] = new_issue.html_url
                        elif res.totalCount == 1:
                            existing_issue = res.get_page(0)[0]
                            new_record[task] = existing_issue.html_url
                        else:
                            print('possible dupes?')
                            for i in res.get_page(0):
                                print('{}'.format(i.html_url))
            new_records.append(new_record)
            rl = gh.get_rate_limit()
            if rl.raw_data['resources']['search']['remaining'] < 20:
                reset_time = rl.raw_data['resources']['search']['reset']
                wait_for = (reset_time - time.time()) + 5
                print('close to rate limit, waiting {} sec'.format(wait_for))
                time.sleep(wait_for)
    with open('data/campaign_finance_new.json','w') as fout:
        json.dump(new_records, fout)
```

```python
new_records = json.load(open('data/campaign_finance_new.json'))
```

```python
updated_records = []
for record in new_records:
    updated = deepcopy(record)
    jurisdiction = record.pop('jurisdiction')
    disclosure_type = record.pop('disclosuretype')
    difficulty = record.pop('difficulty')
    for task, status in record.items():
        if len(status) != "0" and len(status) > 0:
            issue_num = status.split('/')[-1]
            try:
                issue = repo.get_issue(int(issue_num))
            except Exception as e:
                print issue_num
                raise e
            if issue.state == 'closed':
                issue_state = 'done'
            else:
                if issue.assignee is not None:
                    issue_state = 'in-progress'
                else:
                    issue_state = 'todo'

            updated[task] = {'url': issue.html_url, 'state': issue_state}
    updated_records.append(updated)
```
